```python
# replace empty values

NO_IMAGE_URL = "https://upload.wikimedia.org/wikipedia/commons/thumb/a/ac/No_image_available.svg/1024px-No_image_available.svg.png"
df["image.value"] = df["image.value"].fillna(NO_IMAGE_URL)
```

```python
df["image.value"].isna().sum()
```

```python
#Torun borders - let chect if every object is inside borders
#lat: 52.95 – 53.08
#lon: 18.50 – 18.75
min_lat, max_lat = 53.0, 53.1
min_lon, max_lon = 18.55, 18.65

torun_df = df[
    (df['lat.value'] >= min_lat) &
    (df['lat.value'] <= max_lat) &
    (df['lon.value'] >= min_lon) &
    (df['lon.value'] <= max_lon) ]
```

```python
new_df = torun_df[['image.value','itemLabel.value', 'lat.value', 'lon.value']]
```
